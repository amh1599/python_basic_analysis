## 콘텐츠별 수출량

```python
general_data=pd.concat([animation_data,book_data,entertainment_data,cartoon_data,game_data,movie_data,music_data], axis=1)
```

```python
general_data=general_data.astype(str).applymap(lambda x: x.replace(",","")).applymap(lambda x: x.replace("nan","0")).astype("float64").sort_index(ascending=True)
general_data["게임"]=general_data["게임"]/10
general_data["만화"]=general_data["만화"]*10
general_data["영화"]=general_data["영화"]*5
general_data["애니메이션"]=general_data["애니메이션"]*2
```

```python
plt.figure(figsize=(10,7))

sns.pointplot(x=general_data.index, y=general_data["애니메이션"],
              color="red")
sns.pointplot(x=general_data.index, y=general_data["출판"],
              color="pink")
sns.pointplot(x=general_data.index, y=general_data["방송"],
              color="yellow")
sns.pointplot(x=general_data.index, y=general_data["만화"],
              color="green")
sns.pointplot(x=general_data.index, y=general_data["게임"],
              color="skyblue")
sns.pointplot(x=general_data.index, y=general_data["영화"],
              color="purple")
sns.pointplot(x=general_data.index, y=general_data["음악"],
              color="blue")

plt.legend(handles=plt.gca().lines[::len(general_data)+1],labels=["애니메이션", "출판","방송","만화","게임","영화","음악"], loc=2, fontsize=15)
plt.xticks(rotation=30,size=15)
plt.yticks(size=15)
plt.xlabel('년도',size=20)
plt.ylabel('금액',size=20)
plt.title("콘텐츠별 수출액", size= 30)
plt.show()
```
